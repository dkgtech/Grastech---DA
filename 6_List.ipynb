# Nested Tuple

```python
data = (1,2,3,(4,5,6,(7,8,9)))
```

```python
print(data)
```

```python
data = list(data)
print(data)
```

```python
data[3] = list(data[3])
print(data)
```

```python
data[3][3] = list(data[3][3])
print(data)
```

```python
data[3][3][1] = 1000
print(data)
```

```python
data[3][3] = tuple(data[3][3])
print(data)
```

```python
data[3] = tuple(data[3])
print(data)
```

```python
data = tuple(data)
print(data)
```
