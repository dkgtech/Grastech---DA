# Grading System

```python
print(' DPS School '.center(150,'*'))
print()
print()
print()

try:

    maths = float(input('Enter your Maths marks: '))
    hindi = float(input('Enter your Hindi marks: '))
    english = float(input('Enter your English marks: '))
    science = float(input('Enter your Science marks: '))
    sst = float(input('Enter your SST marks: '))

    if (0<=maths<=100) and (0<=hindi<=100) and (0<=english<=100) and (0<=science<=100) and (0<=sst<=100):
        
        marks_obtained = maths+hindi+english+science+sst
        total_marks_paper = 500

        percentage_obtained = round((marks_obtained * 100)/total_marks_paper,2)

        print(f'Total Marks {marks_obtained}')

        if percentage_obtained >=90:
            print(f'You have scored A+ Grade with {percentage_obtained}%')

        elif percentage_obtained >=80:
            print(f'You have scored A Grade with {percentage_obtained}%')

        elif percentage_obtained >=70:
            print(f'You have scored B+ Grade with {percentage_obtained}%')

        elif percentage_obtained >=60:
            print(f'You have scored B Grade with {percentage_obtained}%')

        elif percentage_obtained >=50:
            print(f'You have scored C Grade with {percentage_obtained}%')

        elif percentage_obtained >=40:
            print(f'You have scored D Grade with {percentage_obtained}%')

        elif percentage_obtained >=33:
            print(f'You have scored E Grade with {percentage_obtained}%')

        else:
            print('You are Failed!!')
        
    else:
        print('Incorrect Marks Entered')

except:
    print('Incorrect value Entered')
```

# Number Guessing Game

```python
print(' Number Guess Game '.center(150,'*'))
print()
print()

import random
computer_choice = random.randint(1,6)

try:

    user_choice = int(input('Enter any number between 1 to 6: '))

    if (0<=user_choice<=6):

        print(f'''Your Choice is {user_choice}
    Computer Choice is {computer_choice}''')
        if user_choice == computer_choice:
            print('You Won!! üèÜ')

        else:
            print('Better Luck Next Time!! üòî')

    else:
        print('Enter Number between 1 to 6')

except:
    print('Enter Valid Number')
```

# Rock Paper Scissor Game

```python
print(' Rock Paper Scissor Game '.center(150,'*'))
print()
print()

import random

possible_choices = ['rock','paper','scissor']

comp_choices = random.choice(possible_choices)

us_choice = input('Enter your choice : ').lower()

print(f'''Computer Choice is {comp_choices}
User Choice is {us_choice}''')

if us_choice in possible_choices:

    if comp_choices == us_choice:
        print('Tie : Match Draw')

    elif comp_choices == 'rock':
        if us_choice == 'paper':
            print('User Win and Computer Loose!!')

        else:
            print('Computer Win and User Loose!!')

    elif comp_choices == 'paper':
        if us_choice == 'scissor':
            print('User Win and Computer Loose!!')

        else:
            print('Computer Win and User Loose!!')

    else:
        if us_choice == 'rock':
            print('User Win and Computer Loose!!')

        else:
            print('Computer Win and User Loose!!')

else:
    print('Enter Valid Choice')
```
