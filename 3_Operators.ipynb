# Operators

```python
# Arithmetic Op
# Membership Op
# Comparison Op
# Assignment Op
# Logical Op
# Identity Op
# Bitwise Op
```

```python
# Arithmetic Op

# +
# -
# *
# /
# //
# **
# %

a = 10
b = 20

print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a//b)
print(a**b)
print(a%b)
```

```python
# Comparison Op or Relational Op

# <
# <=
# >
# >=
# ==
# !=

a = 10
b = 20
print(a>b)
print(a>=b)
print(a<b)
print(a<=b)
print(a==b)
print(a!=b)
```

```python
# Comparison Op

# =
# +=
# -=
# *=
# /=
# //=
# **=
# %=

a = 10
print(a)
a += 5
print(a)
a -= 2
print(a)
a *= 10
print(a)
a /= 2
print(a)
a //=2
print(a)
a **=5
print(a)
a %=2
print(a)
```
