# One Line If Else

```python
num = 10

print('Even') if num%2 == 0 else print('Odd')
```

# One Line IF ElIf Else

```python
a = 70
b = 60
c = 80

print('A is the greatest') if (a>b) and (a>c) else print('B is the greatest') if (b>a) and (b>c) else print('C is the greatest') if (c>a) and (c>b) else print('Don\'t Know')
    
```

# Password Generator

```python
import random
import string

pwd_choices = string.ascii_lowercase + string.ascii_uppercase + string.digits + string.punctuation

# print(pwd_choices)

user_want_password_length = 8
password = random.choices(pwd_choices,k = user_want_password_length)

# print(password)

final_password = ''.join(password)
print(final_password)
```

# Shape Details Generator

```python
import IPython

shape = input('Enter the shape name for its detail').lower()

if shape == 'triangle':
    side1 = float(input('Enter 1st Side'))
    side2 = float(input('Enter 2nd Side'))
    side3 = float(input('Enter 3rd Side'))

    if (side1+side2>side3) or (side2+side3>side1) or (side3+side2>side1):
        area = side1*side2*side3
        perimeter = 2*area
        print(f'Area of Triangle is: {area} sq unit')
        print(f'Perimeter of Triangle is: {perimeter} unit')
        display(IPython.display.Image("https://images.rawpixel.com/image_png_800/czNmcy1wcml2YXRlL3Jhd3BpeGVsX2ltYWdlcy93ZWJzaXRlX2NvbnRlbnQvdjExNDgtMDM5LXBfMS1renhoNzVtMy5wbmc.png",height=50,width=50))
       
    
    else:
        print('It is not a Triangle')

elif shape == 'square':
    side = float(input('Enter Side'))
    area = side*side
    perimeter = 4*area
    print(f'Area of Square is: {area} sq unit')
    print(f'Perimeter of Square is: {perimeter} unit')


elif shape == 'rectangle':
    l = float(input('Enter Length'))
    b = float(input('Enter Breadth'))
    area = l*b
    perimeter = 2*(l+b)
    print(f'Area of Rectangle is: {area} sq unit')
    print(f'Perimeter of Rectangle is: {perimeter} unit')


elif shape == 'circle':
    radius = float(input('Enter Radius'))
    pi = 3.14159
    area = (pi*radius*radius)
    perimeter = 2*pi*radius
    print(f'Area of Circle is: {area} sq unit')
    print(f'Perimeter of Circle is: {perimeter} unit')

else:
    print('Unknown Shape')
```
