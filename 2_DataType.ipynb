# Data Types
    # Fundamental Data Types : can store a single value in a object.
        # integer(int) --> will store single value of numerical data with no decimal.
        # float (float) --> will store single value of numerical data with decimal.
        # complex (complex) --> will store single value of numerical data with imaginary part.
        # boolean (bool) --> will store single value of True or False.
        # string (str) --> text data or other than above 3 & in " ".
        
    # Adavance Data Types : can store multiple values in a object.
