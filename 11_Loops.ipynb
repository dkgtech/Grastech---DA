```python
# table print

table = int(input('Enter any number for table printing: '))

for i in range(1,11):
    print(f'{table} x {i} = {table*i}')
```

```python
data = ['Dipesh','Sachin','Rahul','Mahesh','Pritesh']

final_data = []

for i in data:
    final_data.append(i.upper())

print(final_data)
```

```python
data = [2,3,4,4,1323,4343,54543,533,124,24,35,4,556,5,65,65,65,64,5235,325,325,3]

even = []
odd = []

for i in data:
    if i % 2 == 0:
        even.append(i)
    else:
        odd.append(i)

print(even)
print(odd)
```

```python
data = ['Tata','Adani','Reliance','WIPRO','HCL','InFosys']

vowels = []
consonants = []

for i in data:
    if i[0].lower() in 'aeiou':
        vowels.append(i.upper())
    else:
        consonants.append(i.upper())

print(vowels)
print(consonants)
```
