# Number Guess Game

```python
import random
import pygame as pg
import time
pg.init()

limit = 5
company_amount = 0
user_amount = 0

print(' Welcome to the Game '.center(100,'*'))

intro_sound = pg.mixer.Sound(r'/Users/dipeshgupta/Downloads/intro.mp3')
intro_sound.play()

time.sleep(10)

for i in range(1,limit+1):
    company_amount += 10
    comp_choice = random.randint(1,10)
    user_choice = int(input("Enter your choice between 1 to 10"))

    print(f'''Computer Choice: {comp_choice}
    User Choice: {user_choice}''')

    if comp_choice == user_choice:
        print('You Won!! üéá')
        user_amount += 20
        company_amount -= 20
        win_sound = pg.mixer.Sound(r'/Users/dipeshgupta/Downloads/win.mp3')
        win_sound.play()    
    else:
        print('You Loose!! üòî')
        user_amount -= 10
        lost_sound = pg.mixer.Sound(r'/Users/dipeshgupta/Downloads/lost.mp3')
        lost_sound.play()

    print(f'Attempt Left:{limit-i}')
    display(clear= True)

print(f'''Company have: {company_amount}
You have won: {user_amount}''')
```

# Continue

```python
import random
emp_id = [1,2,3,4,5,6,7,8,9]

bonus = []

for i in emp_id:
    if i in (5,6):
        continue # skip
        print('Maza aa gaya')
    else:
        bonus.append((i,random.randint(100,500)))
        print(f'Emp id {i} : <<Bonus Credited>>')

print(bonus)
```
